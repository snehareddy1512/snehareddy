```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Sample product review dataset (e-commerce style reviews)
data = {
    "review": [
        "This product is amazing, exactly what I needed",
        "Terrible quality, broke within a day",
        "Great value for money, highly recommend",
        "Worst purchase ever, complete waste of money",
        "Excellent build quality and fast delivery",
        "Product stopped working after two days",
        "Very happy with this purchase, works perfectly",
        "Poor packaging, item arrived damaged",
        "Amazing product, exceeded my expectations",
        "Not worth the price, very disappointed",
        "Best purchase I made this year, super happy",
        "Bad experience, customer service was rude",
        "Superb quality, will buy again",
        "Completely useless, does not work at all",
        "Loved it, perfect fit and great design",
        "Cheap material, not durable at all",
        "Fantastic product, works as described",
        "Waste of money, regret buying this",
        "Really good product, satisfied with purchase",
        "Horrible experience, product was defective"
    ],
    "sentiment": [
        "positive", "negative", "positive", "negative", "positive",
        "negative", "positive", "negative", "positive", "negative",
        "positive", "negative", "positive", "negative", "positive",
        "negative", "positive", "negative", "positive", "negative"
    ]
}

df = pd.DataFrame(data)
print(df.head())
print("\nTotal reviews:", len(df))
```

```python
# Convert text into numerical features using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["review"])
y = df["sentiment"]

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy*100:.2f}%")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Test the model with new, unseen reviews
def predict_sentiment(review_text):
    vec = vectorizer.transform([review_text])
    prediction = model.predict(vec)[0]
    return prediction

test_reviews = [
    "This is the best product I've ever bought",
    "I regret buying this, total waste of money",
    "Delivery was fast and product quality is great"
]

for review in test_reviews:
    result = predict_sentiment(review)
    print(f"Review: '{review}'  → predicted sentiment: {result}")
```
